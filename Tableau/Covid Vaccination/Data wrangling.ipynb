```python
#read data
vaccinations_doses_by_manuf=pd.read_csv('vaccination-data including type.csv')
vaccinations_doses_by_manuf.head(2)
```

```python
#split vaccines_used columns to individual vaccines type used
df1=vaccinations_doses_by_manuf.loc[:, vaccinations_doses_by_manuf.columns != 'VACCINES_USED']
df2=vaccinations_doses_by_manuf.VACCINES_USED.str.split(',',expand=True).add_prefix('vac_')
vaccinations_doses_by_manuf_expanded=pd.concat([df1, df2], axis=1)
vaccinations_doses_by_manuf_expanded.head(2)
```

```python
vac_cols=['vac_0','vac_1', 'vac_2', 'vac_3', 'vac_4', 'vac_5', 'vac_6',
       'vac_7', 'vac_8']
#creating unique id vac_id using vaccines used
vaccinations_doses_by_manuf_expanded['vac_ID']=vaccinations_doses_by_manuf_expanded[vac_cols].apply(lambda x: ':'.join(x.dropna()), axis=1)
#removing additional "::" generated 
vaccinations_doses_by_manuf_expanded['vac_ID']= vaccinations_doses_by_manuf_expanded['vac_ID'].str.strip(r': {2,}')
vaccinations_doses_by_manuf_expanded.head(2)
```

```python
vaccinations_doses_by_manuf_expanded.to_csv('country_wise_vaccine_corrected.csv')
```
